```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
%pylab inline
```

```python
def plot_trig_function(func):
    n = 10000
    x = np.linspace(-np.pi, np.pi, n)
    plt.figure()
    plt.plot(x, func(x))
    plt.grid()
    plt.setp(plt.gca(), xticklabels=["$-\pi$","$\\frac{-\pi}{2}$",0,"$\\frac{\pi}{2}$" ,"$\pi$"],xticks=(-np.pi,-np.pi/2, 0 ,np.pi/2, np.pi))
    plt.ylim(-2,2)
    title = func.__name__
    plt.title(title)
    plt.savefig(title + ".png")


plot_trig_function(np.sin)
plot_trig_function(np.tan)
def csc(x):
    return 1/np.sin(x)
plot_trig_function(csc)
```
